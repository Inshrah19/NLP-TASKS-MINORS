```python
import pandas as pd
df = pd.read_csv('Study Habits and Activities of Students.csv')


from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource
import numpy as np
output_notebook()
```

```python
#############VISUALIZATION MODE BOKEHJS##################################
```

```python
#BARCHART
stress_counts = df['Stress_Level'].value_counts()
source = ColumnDataSource(data=dict(stress_level=stress_counts.index.astype(str), counts=stress_counts.values))

p_bar = figure(x_range=list(stress_counts.index.astype(str)), title="Stress Level Distribution",
               toolbar_location=None, width=400, height=350)
p_bar.vbar(x='stress_level', top='counts', width=0.7, source=source)
p_bar.xgrid.grid_line_color = None
p_bar.y_range.start = 0
show(p_bar)
```

```python
#HISTOGRAM
hist, edges = np.histogram(df['Sleep_Hours_Per_Day'], bins=10)
p_hist = figure(title="Sleep Hours Per Day Distribution", width=400, height=350)
p_hist.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='navy', line_color='white')
p_hist.xaxis.axis_label = 'Sleep Hours Per Day'
p_hist.yaxis.axis_label = 'Frequency'
show(p_hist)
```

```python

#PLOTLEY VISULAIZATION MODE:
#2nd visulaization mode:

import plotly.express as px
import pandas as pd
df = pd.read_csv('Study Habits and Activities of Students.csv')

# Round Physical Activity Hours to nearest integer as categories
df['Physical_Activity_Rounded'] = df['Physical_Activity_Hours_Per_Day'].round().astype(int)

# Use groupby and reset_index as Plotly works better with DataFrame
activity_data = df.groupby('Physical_Activity_Rounded').size().reset_index(name='Count')

# Define a list of colors to use for the bars, alternating two dark colors repeatedly
colors = ['#8B0000', '#006400']  # Dark Red and Dark Green

# Repeat colors to match the length of data
color_sequence = [colors[i % len(colors)] for i in range(len(activity_data))]

fig_bar = px.bar(
    activity_data,
    x='Physical_Activity_Rounded',
    y='Count',
    title='Physical Activity Hours Distribution (Rounded)',
    color_discrete_sequence=color_sequence
)

fig_bar.show()
```

```python
#histogram
fig_hist = px.histogram(
    df,
    x='Study_Hours_Per_Day',
    nbins=10,
    title='Study Hours Per Day Distribution',
    color_discrete_sequence=['darkorange']
)
fig_hist.show()
```

```python
#scatter graph
import plotly.express as px

# Define a purple-based color sequence (one tone for each category)
purple_colors = ['#800080', '#9932CC', '#8A2BE2']  # Purple, Dark Orchid, Blue Violet

fig = px.scatter(
    df,
    x='Study_Hours_Per_Day',
    y='GPA',
    color='Stress_Level',
    color_discrete_sequence=purple_colors,
    title='Study Hours vs GPA colored by Stress Level',
    labels={'Study_Hours_Per_Day': 'Study Hours Per Day', 'GPA': 'GPA'}
)

fig.show()

```

```python
#Box Plot:
import plotly.express as px

# Define a unique color sequence
custom_colors = ['#FF6347', '#4682B4', '#6A5ACD']  # Tomato red, Steel Blue, Slate Blue

fig = px.box(
    df,
    x='Stress_Level',
    y='GPA',
    title='GPA Distribution by Stress Level',
    labels={'Stress_Level': 'Stress Level', 'GPA': 'GPA'},
    color='Stress_Level',
    color_discrete_sequence=custom_colors
)

fig.show()

```

```python
#Violin Plot
import plotly.express as px

# Define a color map for each Stress_Level category
color_map = {
    'Low': '#8A2BE2',        # Blue Violet
    'Moderate': '#FF69B4',   # Hot Pink
    'High': '#20B2AA'        # Light Sea Green
}

fig = px.violin(
    df,
    x='Stress_Level',
    y='Sleep_Hours_Per_Day',
    box=True,
    points='all',
    color='Stress_Level',
    color_discrete_map=color_map,
    title='Sleep Hours Distribution by Stress Level'
)

fig.show()

```

```python
#Parallel Categories Diagram
df['Physical_Activity_Rounded'] = df['Physical_Activity_Hours_Per_Day'].round().astype(int)

fig = px.parallel_categories(df, dimensions=['Stress_Level', 'Physical_Activity_Rounded'],
                             color='GPA', color_continuous_scale=px.colors.sequential.Inferno,
                             title='Stress Level and Physical Activity Relationship colored by GPA')
fig.show()
```

```python
#Bonus: Correlation and Group Analysis
# Correlation between two numeric columns
# Example: GPA and Study_Hours_Per_Day
df[['GPA', 'Study_Hours_Per_Day']].corr()

# Group by Stress_Level and average GPA
df.groupby('Stress_Level')['GPA'].mean()
```
